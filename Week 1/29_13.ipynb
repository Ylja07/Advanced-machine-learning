```python
import numpy as np
import random
import matplotlib.pyplot as plt

#Function that returns gaussian from given x, mu and sigma
def gauss(x,mu,sigma):
  return (1/np.sqrt(2*np.pi*sigma**2))*np.exp(-(x-mu)**2/(2*sigma**2))

#Variables
mu = 0
sigma1 = 1
sigma2 = np.arange(0.1,1.6,0.05)
R = [1000,10000,100000]
z_vals = []
var_vals = []

#Looping over different values of R
for r in R:
  #Looping over all values of sigma
  for sigma in sigma2:
   np.random.seed(42) #Setting a seed so the plot is continuous for low values of sigma
   x = np.random.normal(mu,sigma,r) #Generating r random normally distributed values
   P = gauss(x,mu,sigma1) #Calculating P from known distribution with sigma = 1
   Q = gauss(x,mu,sigma) #Calculating Q from random different values of sigma
   Z = np.mean(P/Q) #Calculating Z

   #Calculating variance
   var = np.var(P/Q)
   z_vals.append(Z)
   var_vals.append(np.sqrt(var))

#Plotting
plt.plot(sigma2,z_vals[:len(sigma2)],label="1000")
plt.plot(sigma2,z_vals[len(sigma2):2*len(sigma2)],label="10000")
plt.plot(sigma2,z_vals[2*len(sigma2):],label="100000")
plt.plot(sigma2, np.ones(len(sigma2)), label="Theory")
plt.legend()
plt.show()

plt.plot(sigma2,var_vals[:len(sigma2)],label="1000")
plt.plot(sigma2,var_vals[len(sigma2):2*len(sigma2)],label="10000")
plt.plot(sigma2,var_vals[2*len(sigma2):],label="100000")
plt.ylim(0,10)
plt.legend()
plt.show()
```
